```python
# Set path to output file
path = "/Users/shasler/Code/deconfuser_copy/output_files/ten_systems_for_paper/"
lowi_file = os.path.join(path, "11systems_lowi_output_2024-04-02_124415.txt")
```

##### You need the original and ranked dataframes to use some of the plotting functions:

```python
# Ranking to get other dfs
confused_systems = ranking.PhotometryRanking(filepath=lowi_file, n_planets=3) # confused_systems.df gives you original dataframe
df_confused = confused_systems.get_top_group_options()
df_ranked = confused_systems.top_ranked_partition()
df_recombined = confused_systems.combine_and_cleanup()
df_final = confused_systems.orbit_percent_diff()
df_final_wpercdiff = confused_systems.final_recombined()
```

```python
# Set some plotting constants
fig_path = '../path/to/figures'
xlims = [[-5,5], [-3,3], [-3,3], [-5,5], [-3,3], [-1,1], [-5,5], [-3,3], [-6.2,6.2], [-4,4], [-4.1,4.1]]
ylims = [[-5,5], [-3,3], [-3,3], [-5,5], [-4,4], [-1,1], [-3,3], [-3,3], [-6,6], [-4.1,4.1], [-6.5,6.5]]
```

```python
# Plot system detections for just one system
system = 1
lowi_plot = Plot(confused_systems.df, system)                                          # Set up plotting object
lowi_plot.system_detections(plot_tracks=False, save_plot=False,                        # plot_tracks=True if you want orbit tracks 
                            save_path=fig_path+f'lowi_system{system}_detections.png',
                            xlim=xlims[system-1], ylim=ylims[system-1], 
                            plot_title=f'System: {system}', 
                            legend_bbox_anchor=(1.0,1.0))
```

```python
# Plot all confused orbit options per system
lowi_plot.system_options(df_confused, save_plot=False, 
                        save_path=fig_path + f'lowi_system{system}_partition_options.png',
                        xlim=xlims[system-1], ylim=ylims[system-1])
```

```python
# Plot orbit tracks for all confused options
lowi_plot.iterate_over_all_partitions(df_confused, save_plot=False, 
                                      save_path=fig_path, xlims=xlims, ylims=ylims)
```

```python
# Plot top ranked orbits from photometry ranking
lowi_plot.top_partition(df_confused)
```

```python
# Compare top ranked orbits to true simulated orbits for system
lowi_plot.compare_top_to_true(df_ranked, save_plot=False, 
                              save_path=fig_path+f'lowi_system{system}_compare_topranked2sim.png', 
                              xlim=xlims[system-1], ylim=ylims[system-1], 
                              plot_title=True, plot_legend=True) # change to false if you don't want title + legend
```
